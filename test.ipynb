```python
from anneal import SA
import matplotlib.pyplot as plt
import random

import math
from haversine import haversine
```

```python
def read_coords(path):
    coords = []
    with open(path, "r") as f:
        for line in f.readlines():
            line = [float(x.replace("\n", "")) for x in line.split(" ")]
            coords.append(line)
    return coords


def generate_random_coords(num_nodes):
    return [
        [
            random.uniform(-90, 90), random.uniform(-180, 180)
        ] for i in range(num_nodes)
    ]
```

```python
# coords = read_coords("coord.txt") 
coords = generate_random_coords(200)
```

```python
sa = SA(coords, stopping_iter = 50000)
# sa.anneal()
sa.batch_anneal(times = 10)
sa.visualize_routes()
sa.plot_learning()
```

```python
print(sa.best_fitness)
print(sa.best_solution[:5] + ["..."] + sa.best_solution[-5:])
```
